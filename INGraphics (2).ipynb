```python
# We tend to abbreviate the pandas library as pd
import pandas as pd
# Stop pandas from abbreviating tables to fit in the notebook
pd.options.display.max_columns = 1000
pd.options.display.max_rows = 1000
# Display graphs in the notebook
%matplotlib inline
```

```python
# read in our csv file, and automatically change missing values (a dot in the csv) into NaN
boroughs = pd.read_csv('boroughs.csv', na_values = ['.',' '])
# Use the head function to see the amount of rows wanted in the brackets
alls = boroughs.head(len(boroughs))#This means that all the rows are included because the len(boroughs) means the length of boroughs so all of these values will be set to alls
boroughs = boroughs.head(len(boroughs)-5)#These are all the values except UK, England, London, InnerLondon, OuterLondon and these values are set to boroughs
```

```python
boroughs = boroughs.set_index(boroughs.Borough) # This makes the numbers for the first column become replaced by the boroughs instead and sets the values back to boroughs
alls = alls.set_index(alls.Borough) 
```

```python
boroughs = boroughs.sort_values("UnemployY") # This makes the spreadsheet be sorted by the value from unemployed youths instead of being sorted alphabetically by the boroughs
```

```python
boroughs[['InnerOuter','Population','LT15','GCSEs','UnemployY','Jobs','NEET']] #This shows the varibles that we wanted to see from the dataset so it is focused on what we need only
```

```python
# pyplot is the grandparent of all python plotting packages
import matplotlib.pyplot as plt
# seaborn is based on pyplot but makes it easier to use
import seaborn as sns
# Abbreviate seaborn as sns to make easier to write it
```

```python

# by default seaborn plots come out a bit small, so make ours 8in by 8in
plt.figure(figsize=(8,8))
# sns.scatterplot has options for controlling colour and dot size so we can use four variables on one graph
sns.scatterplot(x=boroughs['GCSEs'],
                y=boroughs["UnemployY"],
                size=boroughs["NEET"],
                sizes=(10,200),
                hue=boroughs["LT15"],
                palette="Blues")
# where to put the legend
plt.legend(loc='upper right');
```

```python
boroughs.GCSEs.corr(boroughs.UnemployY) # This shows the correlation coefficient between GCSEs and youth unemployment
```

```python
print(boroughs['GCSEs'].mean())#This is the mean of the GCSE scores added all together from all the boroughs and printed on the screen
print(boroughs['UnemployY'].mean())#This is the mean of the Youth Unemployment scores added all together from all the boroughs and printed on the screen
```
